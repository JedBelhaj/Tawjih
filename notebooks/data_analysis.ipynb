```python
def calculate_t_score(formula, variables):
    if pd.isna(formula):
        return None

    formula = str(formula).strip()

    try:
        return eval(
            formula,
            {
                "__builtins__": {}
            },
            variables,
        )

    except Exception as error:
        print(f"Error in formula: {formula}")
        print(error)

        return None
```

## Your real marks -> FG -> T-score per program -> filter by range/keyword

Subject codes below come from `subject_legend.csv` (F=Français, M=Mathématiques, PH=Philosophie, SP=Sciences Physiques, Ang=Anglais, A=Arabe, Algo=Algorithmique, EP=Éducation Physique, ESP=Espagnol, STI=...).

For subjects you only sat once (Philosophie, Éducation Physique, Espagnol, and Sciences Physiques which shows `99.99`/placeholder in the ADMIS message meaning "not sat in main session"), the same mark is passed as both `main` and `control` to `final_mark`, since `final_mark(x, x) == x` — this keeps a single formula for everything.


```python
def final_mark(main, control=None):
    """Returns the final mark. If control is None, only the main-session mark counts.
    Otherwise: (2*main + control) / 3."""
    if control is None:
        return main
    return (2 * main + control) / 3

# --- Your marks, from the CONTROLE + ADMIS ministry SMS ---
marks = {
    "MG":   final_mark(9.16, 10),      # Moyenne Générale
    "M":    final_mark(7.25, 8),       # Mathématiques
    "Algo": final_mark(8.5, 9.16),     # Algorithmique
    "SP":   final_mark(3.75, 3.75),    # Sciences Physiques (main-session value was the 99.99 placeholder -> ignored)
    "STI":  final_mark(10.4, 12.65),   # Sciences et Technologies de l'Informatique
    "F":    final_mark(8, 6.25),       # Français
    "Ang":  final_mark(15, 17.5),      # Anglais
    "A":    final_mark(9.75, 9.25),    # Arabe
    "PH":   final_mark(6.25, 6.25),    # Philosophie (control session only)
    "EP":   final_mark(18.67, 18.67),  # Éducation Physique (control session only)
    "ESP":  final_mark(13, 13),        # Espagnol (control session only)
}

FG = (
    4 * marks["MG"]
    + 1.5 * marks["M"]
    + 1.5 * marks["Algo"]
    + 0.5 * marks["SP"]
    + 0.5 * marks["STI"]
    + marks["F"]
    + marks["Ang"]
)

variables = {"FG": FG, **marks}
print(f"FG score = {FG:.2f}")
```

### Compute T-score per program

Same `calculate_t_score` as before, but silent: a formula needing a subject you never sat (e.g. `FG+IT` needs Italien) just resolves to `None` instead of printing an error — that program is simply not applicable to you, not a bug.


```python
def calculate_t_score(formula, variables):
    if pd.isna(formula):
        return None
    try:
        return eval(str(formula).strip(), {"__builtins__": {}}, variables)
    except Exception:
        return None  # formula references a subject you didn't sit for

results["T_score"] = results["formula"].apply(lambda f: calculate_t_score(f, variables))
results["T_score"].notna().value_counts()
```

### Search: filter by cutoff range (above/below your score) + keyword across name/institution/university

`above` caps how far a program's cutoff can sit *above* your score (reach schools), `below` caps how far it can sit *below* (safe bets). `keyword` matches against `program_name`, `institution`, or `university` (case-insensitive, Arabic works).


```python
def search_programs(above=10, below=10, keyword=None):
    df = results.dropna(subset=["T_score"]).copy()

    df = df[
        (df["orientation_score"] <= df["T_score"] + above)   # cutoff at most `above` points higher than your score
        & (df["orientation_score"] >= df["T_score"] - below) # cutoff at most `below` points lower than your score
    ]

    if keyword:
        df = df[
            df["program_name"].str.contains(keyword, case=False, na=False)
            | df["institution"].str.contains(keyword, case=False, na=False)
            | df["university"].str.contains(keyword, case=False, na=False)
        ]

    return df[
        ["program_name", "institution", "university", "program_code", "orientation_score", "T_score"]
    ].sort_values("T_score", ascending=False)

# Example: cutoffs up to 5 points above your score (reach) or up to 15 below (safe)
search_programs(above=5, below=15)
```

```python
# Example: same, narrowed to a keyword match in the name/institution/university
search_programs(above=5, below=15, keyword="إعلامية")
```
